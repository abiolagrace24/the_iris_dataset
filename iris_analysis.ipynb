#### Iris Dataset Analysis

###### This notebook walks through a simple data analysis of the famous Iris dataset.  
###### We'll explore how different flower measurements (like petal length and width) vary across species and visualize key patterns.
###### The goal is to practice pandas and visualization basics while learning how to interpret real-world data.


```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
```

```python
# 2. Load the Iris dataset (Excel file)
iris_data = pd.read_csv(r"C:\Users\abiol\OneDrive\Desktop\iris_dataset\Iris.csv")
  # Make sure the file is in your working directory
```

```python
#Preview the dataset to shows the first 5 rows
iris_data.head()
```

```python
#Check for nulls values and data types
iris_data.info()
iris_data.isnull().sum()
```

```python
#Determine the different types of species are there
iris_data['Species'].value_counts()
```

```python
#Visualize petal length by species using a boxplot
sns.boxplot(x='Species', y='PetalLength', data=iris_data)
plt.title("Petal Length by Species")
plt.show()
```

```python
#an error occurred here {Could not interpret value `PetalLength` for `y`. An entry with this name does not appear in `data`.}
#i will be visualizing the iris_data column ti see it the column exists
print(iris_data.columns)
```

##### It was noticed that cm was attached to the end of column name, and i will be removing the suffix cm

```python
# Removing 'Cm' from the end of any column name that contains it
iris_data.columns = iris_data.columns.str.replace('Cm', '', regex=False)

# Print to confirm changes
print(iris_data.columns)
```

```python
#Visualize petal length by species using a boxplot
sns.boxplot(x='Species', y='PetalLength', data=iris_data)
plt.title("Petal Length by Species")
plt.show()
```

```python
#Scatterplot of Petal Length vs Width, colored by species
sns.scatterplot(x='PetalLength', y='PetalWidth', hue='Species', data=iris_data)
plt.title("Petal Length vs Width by Species")
plt.show()
```

```python
#Pairplot of all features in columns shows the relationships between all species all in one look.
sns.pairplot(iris_data, hue='Species')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```
